### Different Data Augmentation techniques

```python
import os
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import skimage.io as io
from skimage.transform import rotate, AffineTransform, warp
from skimage.util import random_noise
from skimage.filters import gaussian
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Add the appropriate folder
#images = os.listdir('images/')
captions = os.listdir('captions_augmented/')

#Select an exampleimage path
image_path = 'all_caption_files/media/images_augmented/01001.jpg'
```

```python
image = io.imread(image_path)
io.imshow(image)
```

```python
#rotating the image by 45 degrees
rotated = rotate(image, angle=90, resize= True, mode = 'symmetric')
#plot the rotated image
io.imshow(rotated)
```

```python
flipLR = np.fliplr(image)

plt.imshow(flipLR)
plt.title('Left to Right Flipped')
```

```python
flipLR.shape
```

```python
flipUD = np.flipud(image)

plt.imshow(flipUD)
plt.title('Up Down Flipped')
```

```python
sigma=0.2
#add random noise to the image
noisyRandom = random_noise(image,var=sigma**2)

plt.imshow(noisyRandom)
plt.title('Random Noise')
```

```python
blurred = gaussian(image,sigma=2,multichannel=True)

plt.imshow(blurred)
plt.title('Blurred Image')
```

```python
from skimage import exposure

#adjusting brightness
image_bright = exposure.adjust_gamma(image, gamma=0.5,gain=1)
image_dark = exposure.adjust_gamma(image, gamma=1.5,gain=1)
```

```python
plt.imshow(image_bright)
```

```python
plt.imshow(image_dark)
```
